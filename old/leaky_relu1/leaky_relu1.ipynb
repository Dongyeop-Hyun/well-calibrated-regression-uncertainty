```python
import torch
from matplotlib import pyplot as plt
import numpy as np
```

```python
def leaky_relu1(x, slope=0.01, a=1):
    x = torch.nn.functional.leaky_relu(x, negative_slope=slope)
    x = -torch.nn.functional.leaky_relu(-x+a, negative_slope=slope)+a
    return x
```

```python
t = torch.linspace(-5,5,1000)
```

```python
plt.plot(t, leaky_relu1(t, 0.1))
plt.grid()
plt.show()
```
